```python
"""Preview the deterministic train/validation/test split used by training."""

from __future__ import annotations

from argparse import ArgumentParser
from collections import Counter
from pathlib import Path

from src.model.data_preparation.loader import DatasetLoader
from src.model.data_preparation.splits import DatasetSplits, SplitBuilder
from src.model.train.validators import DEFAULT_TRAIN_CONFIG_PATH, load_training_config


def build_parser() -> ArgumentParser:
    """Build CLI arguments for the split preview script."""

    parser = ArgumentParser(description="Preview the deterministic data split.")
    parser.add_argument("--train-config", type=Path, default=DEFAULT_TRAIN_CONFIG_PATH)
    parser.add_argument("--dataset-dir", type=Path, default=None)
    parser.add_argument("--validation-size", type=float, default=None)
    parser.add_argument("--test-size", type=float, default=None)
    parser.add_argument("--random-state", type=int, default=None)
    parser.add_argument(
        "--include-synthetic",
        action="store_true",
        help="Include curated synthetic records in the split preview.",
    )
    return parser


def main() -> None:
    """Load data, build splits, and print a compact summary."""

    args = build_parser().parse_args()
    config = load_training_config(
        train_config_path=args.train_config,
        overrides={
            "dataset_dir": args.dataset_dir,
            "validation_size": args.validation_size,
            "test_size": args.test_size,
            "random_state": args.random_state,
            "include_synthetic": args.include_synthetic,
        },
    )
    bundle = DatasetLoader(config.dataset_dir, config.include_synthetic).load()
    splits = SplitBuilder(config).build(bundle)
    print_summary(splits)


def print_summary(splits: DatasetSplits) -> None:
    """Print split sizes and the observed strata in each split."""

    for split_name, records in (
        ("train", splits.train),
        ("validation", splits.validation),
        ("test", splits.test),
        ("provided_other", splits.provided_other),
    ):
        print(f"{split_name}: {len(records)} records")
        if split_name == "provided_other":
            continue
        counts = Counter((record.label, record.length_bucket) for record in records)
        for (label, bucket), count in sorted(counts.items()):
            print(f"  {label}/{bucket}: {count}")
```

```python

if __name__ == "__main__":
    main()
```

```python
import pathlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def count_files_per_class(dataset_root: str) -> dict:
    """
    Walk through the dataset root directory and count .txt files
    in each immediate subdirectory (class folder).
    """
    root = pathlib.Path(dataset_root)
    class_counts = {}
    for subdir in sorted(root.iterdir()):
        if subdir.is_dir():
            # Count .txt files; change the pattern if your files have a different extension
            txt_files = list(subdir.glob("*.txt"))
            class_counts[subdir.name] = len(txt_files)
    return class_counts

def plot_class_distribution(class_counts: dict):
    """
    Create a boxplot of the document counts across classes,
    and a bar chart showing the count for each class.
    """
    # Convert to a DataFrame for easier plotting
    df = pd.DataFrame(list(class_counts.items()), columns=["class", "count"])

    # Boxplot: shows the distribution (min, Q1, median, Q3, max) of class sizes
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df["count"])
    plt.title("Distribution of document counts per class")
    plt.ylabel("Number of documents")
    plt.tight_layout()
    plt.show()

    # Bar chart: exact count for each class
    plt.figure(figsize=(10, 5))
    sns.barplot(data=df, x="class", y="count", palette="viridis")
    plt.xticks(rotation=45, ha="right")
    plt.title("Document count per class")
    plt.ylabel("Number of documents")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Adjust this path to point to your dataset folder
    dataset_path = "dataset"
    counts = count_files_per_class(dataset_path)
    print("Class counts:", counts)
    plot_class_distribution(counts)
```
