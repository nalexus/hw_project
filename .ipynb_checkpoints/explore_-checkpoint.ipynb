#### 1. Exploratory Data Analysis

##### Number of samples per class: investigating class imbalance

Let's first visualize out if we have any class imbalances to know if we need to handle this during modelling. I will define a small helper to count docs per class.

```python
def count_files_per_class(dataset_root: str) -> dict:
    """
    Walk through the dataset root directory and count .txt files
    in each immediate subdirectory (class folder).
    """
    root = pathlib.Path(dataset_root)
    class_counts = {}
    for subdir in sorted(root.iterdir()):
        if subdir.is_dir():
            # Count .txt files;
            txt_files = list(subdir.glob("*.txt"))
            class_counts[subdir.name] = len(txt_files)
    df = pd.DataFrame(list(class_counts.items()), columns=["class", "count"])
    return df

def barplot_wrapper(df, x="class",y="count",title="Document count per class"):
    """
    Create a boxplot of the document counts across classes,
    and a bar chart showing the count for each class.
    """
    # Bar chart: exact count for each class
    plt.figure(figsize=(10, 5))
    sns.barplot(data=df, x=x, y=y, palette="viridis",hue=x,legend=False)
    plt.xticks(rotation=45, ha="right")
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

Let's apply and visualize count_files_per_class results:

```python
barplot_wrapper(count_files_per_class(dataset_path))
```

**Conclusions**: 

- From a very high-level view, we do not have any class imbalance among "known" classes. We do not need to apply any specific sampling techniques or loss function modifications to cover-up specifically known class imbalance during training.
- However, we have a challenge with class "other"/out-of-distribution kind of documents. Technically we don't need to oversample it for training per se, but we still have to recognize this kind of documents during inference stage. This requires from us a robust rejeciton policy, i.e. some kind of threshold which rejects the class with maximum "softmax probability" output. We cannot have enough samples of "other" to make robust estimation of such a rejection policy with this number of "other" samples, so we need to create another way. Possible ideas may be: 1) Some OOD detection techniques like "leave-on-class-out"; 2) Synthetic data generation for "other" etc.

##### Average number unique of tokens per document per class

Let's now analyze how classes differ substantially in number of average normalized tokens per example (a.k.a. "document length"). This finding may direct to various modelling decisions about paramters of preprocessing (TF-IDF) or modelling (transformers), such as type of n-grams or max_token_count used.

```python
def compute_avg_norm_tokens_stats(class_texts: dict[str, list[str]]) -> pd.DataFrame:
    """
    For each class compute average normalized tokens per example
    """
    rows = []

    for cls, texts in class_texts.items():
        texts = [t for t in texts if isinstance(t, str)]
        norm_tokens = []
        for text in texts:
            norm = re.findall(r"\b\w+\b", text.lower())
            norm_tokens.extend(norm)
        n_examples = len(texts)
        n_norm_tokens = len(norm_tokens)
        avg_norm_tokens_per_example = (
            n_norm_tokens / n_examples
            if n_examples > 0
            else 0.0
        )

        rows.append({
            "class": cls,
            "avg_norm_tokens_per_example": avg_norm_tokens_per_example,
        })

    return pd.DataFrame(rows).sort_values("class").reset_index(drop=True)
```

```python
avg_norm_tokens_stats = compute_avg_norm_tokens_stats(class_texts)
```

Now let's see distribution of of average number of normalized tokens per document in each class.

```python
barplot_wrapper(avg_norm_tokens_stats,x='class',y='avg_norm_tokens_per_example',title='Distribution of unique normalized token over classes')
```

The classes differ in average document length. Class "historical" is about 867 normalized tokens per example, while "food", "entertainment", "graphics", "sport", and "business" are mostly around 230-340. 

**Conclusions:**

- Some fixed-context transformer models such as all-MiniLM-L6-v2 with SetFit tuning will almost surely lose important information through truncation. For example, 256 token limit would truncate a large part of "historical", "space", "medical", "politics", and "technologie" documents. Such, chunking, or long-context embeddings should be treated as explicit modelling decisions.
- Given that some classes are much "longer" than others, model performance may depend on length, not only topic semantics. Hupothetically, a model may perform well on short examples but fail on long documents, or vice versa. This may demand using class + length-bucket stratified splitting, and reporting metrics not only by class but also by length bucket.

```python
TOKEN_RE = re.compile(r'\b\w+\b')

def length_bucket(text: str) -> str:
    """Assign a simple word-count bucket for stratification."""
    word_count = len(TOKEN_RE.findall(text))
    if word_count <= 10:
        return "ultra_short"
    if word_count <= 80:
        return "short"
    if word_count <= 300:
        return "medium"
    if word_count <= 700:
        return "long"
    return "extra_long"
    
def compute_vocabulary_stats_by_bucket(class_texts: dict) -> pd.DataFrame:
    """
    For each class and length bucket compute:
      - number of examples
      - unique normalized tokens (lowercased, alphanumeric)
      - average tokens per example
    """
    rows = []
    for cls, texts in class_texts.items():
        # group texts by bucket
        bucket_texts = {}
        for text in texts:
            bucket = length_bucket(text)
            bucket_texts.setdefault(bucket, []).append(text)
        for bucket, bucket_text_list in bucket_texts.items():
            num_examples = len(bucket_text_list)
            all_norm_tokens = []
            total_tokens = 0
            for text in bucket_text_list:
                norm_tokens = TOKEN_RE.findall(text.lower())
                all_norm_tokens.extend(norm_tokens)
                total_tokens += len(norm_tokens)
            unique_norm_tokens = len(set(all_norm_tokens))
            avg_tokens_per_example = total_tokens / num_examples if num_examples else 0.0
            rows.append((cls, bucket, num_examples, unique_norm_tokens, avg_tokens_per_example))
    df = pd.DataFrame(rows, columns=["class", "bucket", "num_examples",
                                     "unique_norm_tokens", "avg_tokens_per_example"])
    # enforce a consistent bucket order for plots
    bucket_order = ["ultra_short", "short", "medium", "long", "extra_long"]
    df["bucket"] = pd.Categorical(df["bucket"], categories=bucket_order, ordered=True)
    df = df.sort_values(["class", "bucket"]).reset_index(drop=True)
    return df
```

```python
vocab_stats_per_bucket = compute_vocabulary_stats_by_bucket(class_texts)
```

```python
import pathlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re



def read_texts_per_class(dataset_root: str) -> dict:
    """
    Read all .txt files in each class folder and return a dict
    mapping class name to list of raw text strings.
    """
    root = pathlib.Path(dataset_root)
    class_texts = {}
    for subdir in sorted(root.iterdir()):
        if subdir.is_dir():
            texts = []
            for txt_path in sorted(subdir.glob("*.txt")):
                try:
                    text = txt_path.read_text(encoding="utf-8")
                except UnicodeDecodeError:
                    # fallback to latin-1 if utf-8 fails
                    text = txt_path.read_text(encoding="latin-1")
                texts.append(text)
            class_texts[subdir.name] = texts
    return class_texts

def token_count_distribution(class_texts: dict):
    """
    Create boxplots of token count and character count per class.
    Token count is computed by splitting on whitespace.
    """
    rows = []
    for cls, texts in class_texts.items():
        for text in texts:
            token_count = len(text.split())
            char_count = len(text)
            rows.append((cls, token_count, char_count))
    df = pd.DataFrame(rows, columns=["class", "token_count", "char_count"])
    return df

import re
import pandas as pd


def compute_vocabulary_stats(class_texts: dict[str, list[str]]) -> pd.DataFrame:
    """
    For each class compute:
      - unique tokens using raw whitespace split
      - unique normalized tokens using lowercased word-like tokens
      - type-token ratio: unique normalized tokens / total normalized tokens
      - average raw whitespace tokens per example
      - average normalized tokens per example
      - number of examples
    """
    rows = []

    for cls, texts in class_texts.items():
        texts = [t for t in texts if isinstance(t, str)]

        raw_tokens = []
        norm_tokens = []

        for text in texts:
            raw = text.split()
            norm = re.findall(r"\b\w+\b", text.lower())

            raw_tokens.extend(raw)
            norm_tokens.extend(norm)

        n_examples = len(texts)
        n_raw_tokens = len(raw_tokens)
        n_norm_tokens = len(norm_tokens)

        unique_raw_tokens = len(set(raw_tokens))
        unique_norm_tokens = len(set(norm_tokens))

        type_token_ratio = (
            unique_norm_tokens / n_norm_tokens
            if n_norm_tokens > 0
            else 0.0
        )

        avg_raw_tokens_per_example = (
            n_raw_tokens / n_examples
            if n_examples > 0
            else 0.0
        )

        avg_norm_tokens_per_example = (
            n_norm_tokens / n_examples
            if n_examples > 0
            else 0.0
        )

        rows.append({
            "class": cls,
            "n_examples": n_examples,
            "n_raw_tokens": n_raw_tokens,
            "n_norm_tokens": n_norm_tokens,
            "unique_raw_tokens": unique_raw_tokens,
            "unique_norm_tokens": unique_norm_tokens,
            "type_token_ratio": type_token_ratio,
            "avg_raw_tokens_per_example": avg_raw_tokens_per_example,
            "avg_norm_tokens_per_example": avg_norm_tokens_per_example,
        })

    return pd.DataFrame(rows).sort_values("class").reset_index(drop=True)

def boxplot_wrapper(df, x='class', y='token_count', title='Token count distribution per class'):
    # Token count boxplot
    plt.figure(figsize=(12, 5))
    sns.boxplot(data=df, x=x, y=y, palette="viridis",hue=x,legend=False)
    plt.xticks(rotation=45, ha="right")
    plt.title(title)
    plt.tight_layout()
    plt.show()
                    
if __name__ == "__main__":
    # Adjust this path to point to your dataset folder
    dataset_path = "data/dataset"
    
    df = count_files_per_class(dataset_path)
    barplot_wrapper(df)
    
    # Text length analysis
    class_texts = read_texts_per_class(dataset_path)
    token_df = token_count_distribution(class_texts)
    boxplot_wrapper(token_df)

    # Vocabulary size analysis
    vocab_df = compute_vocabulary_stats(class_texts)
    barplot_wrapper(vocab_df,'class','unique_norm_tokens',"Unique normalized tokens per class")
```
